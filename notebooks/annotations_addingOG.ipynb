```python
import os
import pandas as pd
```

```python
annotations_dir = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_namemaps/annotations_conversions/"
tx2gene_dir = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_salmon/tx2gene/"
annotations_files = os.listdir(annotations_dir)
tx2gene_files = os.listdir(tx2gene_dir)
```

```python
annotation_file = annotations_dir + "F_rathbuni.annotations.conversions.csv"
annotations = pd.read_csv(annotation_file,index_col=0)
```

```python
annotations.head()
```

```python
tx2gene_file = tx2gene_dir + "F_rathbuni_tx2gene.csv"
tx2gene = pd.read_csv(tx2gene_file,index_col=0)
```

```python
tx2gene.head()
```

```python
tx2gene['extra'], tx2gene['TrinityID'] = tx2gene['transcriptID'].str.split('-', 1).str
tx2gene = tx2gene.drop(columns=['extra'])
```

```python
tx2gene.head()
```

```python
# split NCBI
annotations['split1'], annotations['split2'],annotations['split3'] = annotations['NCBI'].str.split('|', 3).str
annotations = annotations.drop(columns=['split1'])
annotations = annotations.rename(columns = {'split2':'NCBI_protein','split3':'NCBI_name'})
#annotations = annotations.drop(columns=['NCBI'])
```

```python
# split evigene
annotations['split1'], annotations['split2'] = annotations['EviGene'].str.split('t', 2).str
annotations = annotations.drop(columns=['split2'])
annotations = annotations.rename(columns = {'split1':'EviGene_gene','EviGene':'EviGene_transcript'})
```

```python
# split Ensembl
annotations['split1'], annotations['split2'] = annotations['Ensembl'].str.split(' ', 1).str
annotations = annotations.drop(columns=['split2'])
annotations = annotations.rename(columns = {'split1':'Ensembl','Ensembl':'Ensembl_info'})
```

```python
annotations.head()
```

```python
merged = pd.merge(tx2gene,annotations,on='TrinityID',how='outer')
merged = merged.sort_values(by=['seqid'])
```

```python
merged.head()
```

# all species

```python
for annotation_file in annotations_files:
    species = annotation_file.split(".")[0]
    annotation_file = annotations_dir + annotation_file
    annotations = pd.read_csv(annotation_file,index_col=0)
    tx2gene_file = tx2gene_dir + species + "_tx2gene.csv"
    tx2gene = pd.read_csv(tx2gene_file,index_col=0)
    tx2gene['extra'], tx2gene['TrinityID'] = tx2gene['transcriptID'].str.split('-', 1).str
    tx2gene = tx2gene.drop(columns=['extra'])
    # split NCBI
    annotations['split1'], annotations['split2'],annotations['split3'] = annotations['NCBI'].str.split('|', 3).str
    annotations = annotations.drop(columns=['split1'])
    annotations = annotations.rename(columns = {'split2':'NCBI_protein','split3':'NCBI_name'})
    # split evigene
    annotations['split1'], annotations['split2'] = annotations['EviGene'].str.split('t', 2).str
    annotations = annotations.drop(columns=['split2'])
    annotations = annotations.rename(columns = {'split1':'EviGene_gene','EviGene':'EviGene_transcript'})
    # split Ensembl
    annotations['split1'], annotations['split2'] = annotations['Ensembl'].str.split(' ', 1).str
    annotations = annotations.drop(columns=['split2'])
    annotations = annotations.rename(columns = {'split1':'Ensembl','Ensembl':'Ensembl_info'})
    merged = pd.merge(tx2gene,annotations,on='TrinityID',how='outer')
    merged = merged.sort_values(by=['seqid'])
    new_merged_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_namemaps/merged_annotations/" + species + ".annotations.OGs.merged.csv"
    merged.to_csv(new_merged_file)
    print("Written:",new_merged_file)
```
