#  Connecting counts with annotations

```python
import os
import pandas as pd
```

# all species

```python
annotations_dir = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_expression_July2019/merged_annotations/"
annotations_files = os.listdir(annotations_dir)
```

```python
print(annotations_files)
```

```python
for annotation_file in annotations_files:
    species = annotation_file.split(".")[0]
    print("===============")
    print(species)
    print("===============")
    annotation_file = annotations_dir + annotation_file
    annotations = pd.read_csv(annotation_file,index_col=0)
    # make NCBI
    ncbi_tx2gene = annotations[['TrinityID','NCBI_protein']]
    #ncbi_tx2gene = ncbi_tx2gene.fillna(value={'NCBI_protein':''})
    ncbi_tx2gene = ncbi_tx2gene.dropna()
    print('NCBI:',ncbi_tx2gene['NCBI_protein'].nunique())
    tx2gene_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_expression_July2019/tx2gene_ncbi/" + species + ".ncbi.tx2gene.csv"
    ncbi_tx2gene.to_csv(tx2gene_file)
    print("Written:",tx2gene_file)
    # make Evigene
    evigene_tx2gene = annotations[['TrinityID','EviGene_gene']]
    #evigene_tx2gene = evigene_tx2gene.fillna(value={'EviGene_gene':''})
    evigene_tx2gene = evigene_tx2gene.dropna()
    print('Evigene',evigene_tx2gene['EviGene_gene'].nunique())
    tx2gene_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_expression_July2019/tx2gene_evigene/" + species + ".evigene.tx2gene.csv"
    evigene_tx2gene.to_csv(tx2gene_file)
    print("Written:",tx2gene_file)
    # make Ensembl
    ensembl_tx2gene = annotations[['TrinityID','Ensembl']]
    #ensembl_tx2gene = ensembl_tx2gene.fillna(value={'Ensembl':''})
    ensembl_tx2gene = ensembl_tx2gene.dropna()
    print('Ensembl',ensembl_tx2gene['Ensembl'].nunique())
    tx2gene_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/kfish_expression_July2019/tx2gene_ensembl/" + species + ".ensembl.tx2gene.csv"
    ensembl_tx2gene.to_csv(tx2gene_file)
    print("Written:",tx2gene_file)
```
