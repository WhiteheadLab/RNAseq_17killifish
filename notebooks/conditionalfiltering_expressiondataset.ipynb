# experiment-specific filtering regime

Require at least one sample from one marine species from each clade. This will capture genes that are conditionally turned off and also capture what marine species are doing that freshwater species are no longer doing, because they've lost hyperosmotic regulation

OR

Check if it's on at least in Freshwater species in clade.
If not, then check marine.
This will get genes that are representative from each physiology and allow for comparison with corresponding responses in the physiological counterpart, within each clade.

```python
# nonzero counts in at least two samples in at least one freshwater species within a clade
# if absent in at least one freshwater species, per clade
# go to the marine and check for at least two samples in at least one marine
# vice versa
# loop through index (samples) each dataframe:
# M_Clade1.index.values.tolist()
# they are grouped by clade and physiology
# now need to loop through species in clade dataframe
# species_clade = df['species'].unique()
#FW_Clade1
#M_Clade2
#FW_Clade2
#M_Clade3
#FW_Clade3
# checking for nonzero expression in counts dataframe for those columns
# then merge all together
```

# Clade

```python
M_Clade1 = design.loc[design['phys_clade'] == 'M_Clade1']
M_Clade1.shape
```

```python
FW_Clade1 = design.loc[design['phys_clade'] == 'FW_Clade1']
FW_Clade1.shape
```

```python
FW_Clade2 = design.loc[design['phys_clade'] == 'FW_Clade2']
FW_Clade2.shape
```

```python
M_Clade2 = design.loc[design['phys_clade'] == 'M_Clade2']
M_Clade2.shape
```

```python
M_Clade3 = design.loc[design['phys_clade'] == 'M_Clade3']
M_Clade3.shape
```

```python
FW_Clade3 = design.loc[design['phys_clade'] == 'FW_Clade3']
FW_Clade3.shape
```

```python
FW_Clade3.head()
```

```python
samples_clade = M_Clade1.index.values.tolist()
new_df = counts[counts.columns.intersection(samples_clade)]
new_df = new_df.apply(pd.to_numeric)
#print(new_df.head())
print(new_df.shape)
nonzero = new_df[(new_df != 0.0).any(1)]
print(nonzero.shape)
```

```python
# loop through index (samples) each dataframe,
cladedfs = [M_Clade1,
            FW_Clade1,
            M_Clade2,
            FW_Clade2,
            M_Clade3,
            FW_Clade3]
newdfs = []
for i in cladedfs:
    samples_clade = i.index.values.tolist()
    # subset reads based on samples in each dataframe
    # df.columns.intersection(lst)
    new_df = counts[counts.columns.intersection(samples_clade)]
    # because data types matter
    new_df = new_df.apply(pd.to_numeric)
    # drop gene/row where there are any zeros
    #nonzero = new_df[(new_df != 0).all(1)]
    nonzero = new_df[(new_df != 0.0).any(1)]
    #print(nonzero.columns)
    #print(nonzero.head())
    newdfs.append(nonzero)

# they are grouped by clade and physiology
# now need to loop through species in clade dataframe
# species_clade = df['species'].unique()
```

```python
from functools import reduce
merged = reduce(lambda x, y: pd.merge(x, y, on = 'Ensembl'), newdfs)
merged.shape
```

# Species

Now, subset counts for species

```python
# split design by species
species_list=["A_xenica","F_catanatus","F_chrysotus","F_diaphanus","F_grandis",
"F_heteroclitusMDPL","F_heteroclitusMDPP","F_notatus","F_olivaceous",
"F_parvapinis","F_rathbuni","F_sciadicus","F_similis","L_goodei","L_parva"]
species_design_list = []
for species in species_list:
    new_design = design.loc[design['species'] == species]
    species_design_list.append(new_design)
```

```python
speciesdfs = []
for species_design in species_design_list:
    samples_species = species_design.index.values.tolist()
    # subset reads based on samples in each dataframe
    # df.columns.intersection(lst)
    new_df = counts[counts.columns.intersection(samples_species)]
    nonzero = new_df[(new_df != 0).any(axis=1)]
    speciesdfs.append(nonzero)
```

```python
merged = reduce(lambda x, y: pd.merge(x, y, on = 'Ensembl'), speciesdfs)
merged.shape
```

# Add design rows

```python
pd.set_option('display.max_columns',500)
merged.head()
```

```python
cols = merged.columns.tolist()
```

```python
clade_design = {"A_xenica":"Clade3","F_catanatus":"Clade1","F_chrysotus":"Clade3",
 "F_diaphanus":"Clade1","F_grandis":"Clade1","F_heteroclitusMDPL":"Clade1",
 "F_heteroclitusMDPP":"Clade1","F_notatus":"Clade3","F_notti":"Clade3",
 "F_olivaceous":"Clade3","F_parvapinis":"Clade2","F_rathbuni":"Clade1",
 "F_sciadicus":"Clade3","F_similis":"Clade1","F_zebrinus":"Clade3",
 "L_goodei":"Clade2","L_parva":"Clade2"}
combined_design = {"A_xenica":"M_Clade3","F_catanatus":"FW_Clade1","F_chrysotus":"BW_Clade3",
 "F_diaphanus":"BW_Clade1","F_grandis":"M_Clade1","F_heteroclitusMDPL":"M_Clade1",
 "F_heteroclitusMDPP":"M_Clade1","F_notatus":"FW_Clade3","F_notti":"FW_Clade3",
 "F_olivaceous":"FW_Clade3","F_parvapinis":"BW_Clade2","F_rathbuni":"FW_Clade1",
 "F_sciadicus":"FW_Clade3","F_similis":"M_Clade1","F_zebrinus":"M_Clade3",
 "L_goodei":"FW_Clade2","L_parva":"M_Clade2"}
phys_design = {"A_xenica":"M","F_catanatus":"FW","F_chrysotus":"BW",
 "F_diaphanus":"BW","F_grandis":"M","F_heteroclitusMDPL":"M",
 "F_heteroclitusMDPP":"M","F_notatus":"FW","F_notti":"FW",
 "F_olivaceous":"FW","F_parvapinis":"BW","F_rathbuni":"FW",
 "F_sciadicus":"FW","F_similis":"M","F_zebrinus":"M",
 "L_goodei":"FW","L_parva":"M"}
```

```python
species_list = []
physiology_list = []
clade_list = []
combined_list = []
condition_list = []
for col in cols:
    if col != "OG":
        species = col.split("_")[0]+"_"+col.split("_")[1]
        condition = col.split("_")[2]
        if condition == "BW":
            condition = "15_ppt"
        if condition == "FW":
            condition = "0.2_ppt"
        condition_list.append(condition)
        species_list.append(species)
        physiology = phys_design[species]
        physiology_list.append(physiology)
        clade = clade_design[species]
        clade_list.append(clade)
        design = combined_design[species]
        combined_list.append(design)
print(len(species_list))
print(len(physiology_list))
print(len(clade_list))
print(len(combined_list))
print(len(condition_list))
```

```python
merged.columns
```

```python
df = pd.DataFrame([species_list],columns=merged.columns)
df2 = pd.DataFrame([physiology_list],columns=merged.columns)
df2 = df2.replace('BW', 'M')
df3 = pd.DataFrame([clade_list],columns=merged.columns)
df4 = pd.DataFrame([combined_list],columns=merged.columns)
df5 = pd.DataFrame([condition_list],columns=merged.columns)
print(df.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
print(df5.shape)
df.head()
```

```python
species_merged = merged.append(df)
species_merged = species_merged.append(df2)
species_merged = species_merged.append(df3)
species_merged = species_merged.append(df4)
species_merged = species_merged.append(df5)
print(species_merged.shape)
```
