# experiment-specific filtering regime

```python
import pandas as pd
from functools import reduce
```

```python
# load counts dataframe
data_file = "/Users/johnsolk/Documents/UCDavis/Whitehead/Ensembl_species_counts_designfactors.csv"
data = pd.read_csv(data_file,index_col='Ensembl',low_memory=False)
```

```python
counts = data[:-5]
counts = counts.drop(['Unnamed: 0'],axis=1)
counts.shape
```

```python
counts.head()
```

```python
data.shape
```

```python
design = data[-5:]
design = design.drop(['Unnamed: 0'],axis=1)
```

```python
design = design.T
design.head()
```

```python
design.columns = ['species','physiology','clade','phys_clade','condition']
design = design.replace('BW', 'M')
design = design.replace('BW_Clade3', 'M_Clade3')
design = design.replace('BW_Clade2', 'M_Clade2')
design = design.replace('BW_Clade1', 'M_Clade1')
```

# Clade

```python
M_Clade1 = design.loc[design['phys_clade'] == 'M_Clade1']
M_Clade1.shape
```

```python
FW_Clade1 = design.loc[design['phys_clade'] == 'FW_Clade1']
FW_Clade1.shape
```

```python
FW_Clade2 = design.loc[design['phys_clade'] == 'FW_Clade2']
FW_Clade2.shape
```

```python
M_Clade2 = design.loc[design['phys_clade'] == 'M_Clade2']
M_Clade2.shape
```

```python
M_Clade3 = design.loc[design['phys_clade'] == 'M_Clade3']
M_Clade3.shape
```

```python
FW_Clade3 = design.loc[design['phys_clade'] == 'FW_Clade3']
FW_Clade3.shape
```

```python
FW_Clade3.head()
```

```python
samples_clade = M_Clade1.index.values.tolist()
new_df = counts[counts.columns.intersection(samples_clade)]
new_df = new_df.apply(pd.to_numeric)
#print(new_df.head())
print(new_df.shape)
nonzero = new_df[(new_df != 0.0).any(1)]
print(nonzero.shape)
greaterthan = nonzero[(nonzero > 5.0).any(1)]
print(greaterthan.shape)
```

```python
# loop through index (samples) each dataframe,
cladedfs = [M_Clade1,
            FW_Clade1,
            M_Clade2,
            FW_Clade2,
            M_Clade3,
            FW_Clade3]
newdfs = []
for i in cladedfs:
    samples_clade = i.index.values.tolist()
    # subset reads based on samples in each dataframe
    # df.columns.intersection(lst)
    new_df = counts[counts.columns.intersection(samples_clade)]
    # because data types matter
    new_df = new_df.apply(pd.to_numeric)
    # drop gene/row where there are any zeros
    #nonzero = new_df[(new_df != 0).all(1)]
    nonzero = new_df[(new_df != 0.0).any(1)]
    nonzero = nonzero.apply(pd.to_numeric)
    greaterthan = nonzero[(nonzero > 5.0).any(1)]
    #print(nonzero.columns)
    #print(nonzero.head())
    newdfs.append(greaterthan)

# they are grouped by clade and physiology
# now need to loop through species in clade dataframe
# species_clade = df['species'].unique()
```

```python
merged = reduce(lambda x, y: pd.merge(x, y, on = 'Ensembl'), newdfs)
print(merged.shape)
merged = merged.apply(pd.to_numeric)
merged_nonzero = merged[(merged != 0).all(1)]
print(merged_nonzero.shape)
```

# Species

Now, subset counts for species

```python
# split design by species
species_list=["A_xenica","F_catanatus","F_chrysotus","F_diaphanus","F_grandis",
"F_heteroclitusMDPL","F_heteroclitusMDPP","F_notatus","F_olivaceous",
"F_parvapinis","F_rathbuni","F_sciadicus","F_similis","L_goodei","L_parva"]
species_design_list = []
for species in species_list:
    new_design = design.loc[design['species'] == species]
    species_design_list.append(new_design)
```

```python
# mean of individuals in species,condition >1
```

```python
speciesdfs = []
for species_design in species_design_list:
    samples_species = species_design.index.values.tolist()
    # collect samples based on condition
    # throw out transfer (for now)
    BW = []
    FW = []
    for sample in samples_species:
        condition = sample.split("_")[2]
        if condition == "BW":
            BW.append(sample)
        elif condition == "FW":
            FW.append(sample)
    # subset reads based on samples in each BW and FW dataframe
    # df.columns.intersection(lst)
    print(BW)
    print(FW)
    BW_df = counts[counts.columns.intersection(BW)]
    FW_df = counts[counts.columns.intersection(FW)]
    # because data types matter
    BW_df = BW_df.apply(pd.to_numeric)
    FW_df = FW_df.apply(pd.to_numeric)
    print(BW_df.shape)
    # require mean > 1 of each BW and FW
    BW_mean = BW_df[BW_df.mean(axis=1)>1]
    FW_mean = FW_df[FW_df.mean(axis=1)>1]
    print(BW_mean.shape)
    speciesdfs.append(BW_mean)
    speciesdfs.append(FW_mean)
```

```python
merged = reduce(lambda x, y: pd.merge(x, y, on = 'Ensembl'), speciesdfs)
merged.shape
```
