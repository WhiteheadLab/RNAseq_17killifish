```python
presence_absence = pd.DataFrame()
presence_absence['Ensembl'] = ""  
for genes_file in counts_matrices:
    if genes_file.endswith(".csv"):
        species = genes_file.split("_")[0]+"_"+genes_file.split("_")[1]
        print(species)
        table = pd.read_csv(counts_dir+genes_file,sep="\t")
        table = table.rename(columns={'Unnamed: 0': 'Ensembl'})
        table_filt = table[(table.iloc[:,1:] > 5).any(1)]
        print(table.shape)
        #print(table_filt.shape)
        colnames = ['Ensembl',species]
        species_presence = pd.DataFrame()
        species_presence['Ensembl'] = table['Ensembl']
        species_presence[species] = '1'
        presence_absence = pd.merge(presence_absence,species_presence,on='Ensembl',how='outer')
        print(presence_absence.shape)
        # construct unique gene list ()
```

```python
presence_absence = presence_absence.fillna(0)
```

```python
grouped_d = {}
for spname in presence_absence.columns:
    if spname == 'Ensembl':
        continue
    sp_df = pd.DataFrame(presence_absence, columns=["Ensembl", spname])
    sp_df = sp_df.loc[sp_df[spname] == '1']
    print(spname, len(sp_df))
    grouped_d[spname] = sp_df
```

```python
keep = ['F_sciadicus', 'F_rathbuni', 'F_zebrinus', 'F_grandis', 'F_notatus',
       'F_parvapinis','L_goodei','F_olivaceous','L_parva','F_heteroclitusMDPP',
       'F_similis','F_diaphanus','F_chrysotus','F_nottii','A_xenica','F_catanatus','F_heteroclitusMDPL']

keylist = list(grouped_d.keys())
for k in keylist:
    if k not in keep:
        del grouped_d[k]
```
